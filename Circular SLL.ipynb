# Circular SLL

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    #insertion node at beginning   
    def add_beg(self,data):
        nb = Node(data)
        if self.head is None:
            self.head = nb
            self.tail = nb
            self.tail.next = self.head
        else:
            nb.next = self.head
            self.tail.next = nb
            self.head = nb
            
    #insertion node at ending    
    def add_end(self,data):
        ne = Node(data)
        if self.head is None:
            self.head = ne
            self.tail = ne
            self.tail.next = self.head
        else:
            self.tail.next = ne
            self.tail = ne
            self.tail.next = self.head
            
    #insertion node at specific position         
    def add_pos(self,pos,data):
        np = Node(data)
        if self.head is None:
            self.head = np
            self.tail = np
            self.tail.next = self.head
        else:
            temp = self.head
            for i in range(0,pos-1):
                temp = temp.next
            np.next = temp.next
            temp.next = np  
                
    #Deleting node at beginning            
    def del_beg(self):
        if self.head is None:
            print("CLL is Empty")
        else:
            temp = self.head
            self.head = temp.next
            temp.next = None
            self.tail.next = self.head
                
    #Deleting node at ending
    def del_end(self):
        if self.head is None:
            print("CLL is Empty")
        else:
            temp = self.head
            while temp.next != self.tail:
                temp = temp.next
            self.tail.next = None
            self.tail = temp
            temp.next = self.head
                
    #Deleting node at specific position            
    def del_pos(self,pos):
        if self.head is None:
            print("CLL is Empty")
        else:
            temp1 = self.head
            temp2 = self.head.next
            for i in range(pos-1):
                temp1 = temp1.next
                temp2 = temp2.next
            temp1.next = temp2.next
            if temp2 == self.tail:
                self.tail = temp1
            temp2 = None  
            
    def display(self):
        temp = self.head
        if temp is None:
            print("CLL is Empty")
        else:
            temp = self.head
            print(temp.data,"-->",end=" ")
            while temp.next != self.head:
                temp = temp.next
                print(temp.data,"-->",end=" ")
            print(temp.next.data)

L = CSLL()
n1 = Node(10)
L.head = n1
L.tail = n1
L.tail.next = n1
print("After 1st Node:")
L.display()

n2 = Node(20)
L.tail.next = n2
L.tail = n2
L.tail.next = L.head
print("\nAfter 2st Node:")
L.display()

n3 = Node(30)
L.tail.next = n3
L.tail = n3
L.tail.next = L.head
print("\nAfter 3rd Node:")
L.display()

print("\nAfter inserting node at beginning :")
L.add_beg(5)
L.display()

print("\nAfter inserting node at ending :")
L.add_end(40)
L.display()

print("\nAfter inserting node at specific position :")
L.add_pos(2,15)
L.display()

print("\nAfter deleting node at beginning :")
L.del_beg()
L.display()

print("\nAfter deleting node at ending :")
L.del_end()
L.display()

print("\nAfter deleting node at specific position :")
L.del_pos(2)
L.display()
```

```python
#Searching an Element

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def search(self):
        x = int(input("enter element to be searched in CLL: "))
        temp = self.head
        count = 0
        flag = 0
        while temp.next:
            if x == temp.data:
                flag = 1
                break
            temp = temp.next
            count = count+1
        if flag == 1:
            print(x," is found at position ",count)
        else:
            print(x," is not found in CLL")
        
    def display(self):
        temp = self.head
        if temp is None:
            print("CLL is Empty")
        else:
            temp = self.head
            print(temp.data,"-->",end=" ")
            while temp.next != self.head:
                temp = temp.next
                print(temp.data,"-->",end=" ")
            print(temp.next.data)
            
L = CSLL()
n1 = Node(10)
L.head = n1
L.tail = n1
L.tail.next = n1

n2 = Node(20)
L.tail.next = n2
L.tail = n2
L.tail.next = L.head

n3 = Node(30)
L.tail.next = n3
L.tail = n3
L.tail.next = L.head

n4 = Node(40)
L.tail.next = n4
L.tail = n4
L.tail.next = L.head

L.display()

L.search()
```
